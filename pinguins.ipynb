```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tabela = pd.read_csv('pinguins.csv',  delimiter=';')
```

```python
tabela.head()
```

```python
tabela.info()
```

```python
# Verificar dados ausentes
missing_data = tabela.isnull().sum()
print(missing_data)
```

```python
# Pinguins com dados ausentes
penguins_missing = tabela[tabela.isnull().any(axis=1)]
print(penguins_missing)
```

```python
# Contar pinguins por ilha
island_counts = tabela['Ilha'].value_counts()
print(island_counts)
```

```python
# Contar pinguins por esp√©cie
species_counts = tabela['Espece'].value_counts()
print(species_counts)
```

```python
# Pairplot para explorar rela√ß√µes entre medidas e esp√©cies
sns.pairplot(tabela, hue='Espece')
plt.show()
```

```python
# FacetGrid para explorar rela√ß√µes entre medidas e sexo dentro de cada esp√©cie
g = sns.FacetGrid(tabela, col="Espece", hue="sexo", height=4, aspect=1)
g.map_dataframe(sns.scatterplot, x='profundidade do bico', y='profundidade do bico.1')
g.add_legend()
plt.show()
```

```python
g = sns.FacetGrid(tabela, col="Espece", hue="sexo", height=4, aspect=1)
g.map_dataframe(sns.scatterplot, x='largura do bico', y='massa corporal')
g.add_legend()
plt.show()
```
